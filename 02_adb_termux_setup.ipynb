## Troubleshooting

### Device Shows as `unauthorized`

Example:

```
List of devices attached
XXXXXXXXXXXX    unauthorized
```

Cause:

* USB debugging authorization has not been accepted on the Android device.

Resolution:

1. Disconnect and reconnect the USB cable.
2. Unlock the Android device.
3. Accept the USB debugging prompt.
4. Run:

```bash
adb devices
```

again.

If the issue persists:

```bash
adb kill-server
adb start-server
adb devices
```

Then reconnect the device and accept the authorization prompt.

---

### Device Not Detected

Example:

```
List of devices attached
```

with no devices listed.

Possible Causes:

* USB debugging is disabled.
* USB cable supports charging only.
* Device drivers are not installed.
* USB connection mode is incorrect.

Resolution:

1. Enable Developer Options.
2. Enable USB Debugging.
3. Set USB mode to File Transfer (MTP).
4. Try a different USB cable.
5. Run:

```bash
adb devices
```

again.

---

### Verify ADB Connectivity

A successful connection should appear as:

```

```

The device is now ready for file transfer and deployment.

