## 4. Verify model exists locally

```python
from pathlib import Path

possible_models = list(Path(".").glob("*.gguf"))

if possible_models:
    model = possible_models[0]

    print("Model found:")
    print(model)
    print(f"Size: {model.stat().st_size / 1024**3:.2f} GB")
else:
    print(
        "No GGUF model found. Run "
        "01_hf_to_gguf_q4_0.ipynb first."
    )
```
